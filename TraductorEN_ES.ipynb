```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

def translate_with_transformers(text):
    model_name = "Helsinki-NLP/opus-mt-en-es"
    # Cargamos modelo y tokenizador
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    # Tokenizamos el texto de entrada
    inputs = tokenizer(text, return_tensors="pt", padding=True)

    # Generamos la traducción directamente con el modelo
    translated_tokens = model.generate(**inputs)

    # Decodificamos los tokens a texto
    result = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    return result
```

```python
# Función principal para elegir la traducción
def translate(text, method="transformers"):
    if method == "transformers":
        return translate_with_transformers(text)
    else:
        return "Método de traducción no válido."
```

```python
# Ejemplo de uso
text = """I had a great weekend. On Friday afternoon, I finished work at 5 PM.\nI went home and took a shower. Then I went to see a couple of my friends at a bar downtown."""
try:
    # Ejecuta D5-KHWDpgFEP y u25icQrFgmYs antes que esta celda
    traduccion = translate(text, method="transformers")
    print("Traducción directa con Modelo:\n", traduccion)
except Exception as e:
    import traceback
    print(f"Error al ejecutar la traducción: {e}")
    traceback.print_exc()
```
